```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

  ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
len(toy_dataset)
```

```python
# Apply one-hot encoding to the dataset
te = TransactionEncoder()
transform = te.fit(toy_dataset).transform(toy_dataset)
df = pd.DataFrame(transform, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

```python
frequent_itemsets = apriori(df,min_support=0.6,use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)#associate items
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

# Run Apriori Algorithm on Checkpoint Dataset

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#Convert the dataset to the required format using one-hot encoding
te = TransactionEncoder()
te_ary = te.fit(data_list).transform(data_list)
data_transformed = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Run the Apriori algorithm
frequent_itemsets2 = apriori(data_transformed, min_support=0.02, use_colnames=True)
```

```python
# Extract association rules
rules2 = association_rules(frequent_itemsets2, metric="lift", min_threshold=1)
```
