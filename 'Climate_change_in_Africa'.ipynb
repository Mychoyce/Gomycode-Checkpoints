Ploting a line chart to show the average temperature fluctuations in Tunisia and Cameroon

```python
#to adjust the line of the date of the from 1980-2005
df2 = df[df["DATE"]<'2006']
```

```python
#To adjust the line chart to show dataset for tunisa and cameroon
df3 = df2[(df2["COUNTRY"] =='Tunisa') | (df2["COUNTRY"] == 'Cameroon')]
```

```python
#code to draw a line chart to show dataset for tunsia and cameroon
fig= px.line(df2, x="DATE",y="TAVG",title= 'Average temperature fluctuation in Tunsia and Cameroon')
```

```python
#show the line chart
fig.show()
```
